**Z_score outlier detection for Daily_Usage_Hours column**

```python
Daily_Usage_Hours_mean=df['Daily_Usage_Hours'].mean()
```

```python
Daily_Usage_Hours_std_dev=df['Daily_Usage_Hours'].std()
```

```python
df['Daily_Usage_Hours']
```

```python
df['z_score_Daily_Usage_Hours']=(df['Daily_Usage_Hours']-Daily_Usage_Hours_mean)/Daily_Usage_Hours_std_dev
```

```python
threshold=3
outliers_Daily_Usage_Hours=df[np.abs(df['z_score_Daily_Usage_Hours']>threshold)]
```

```python
outliers_Daily_Usage_Hours
```

**capping**

```python
outlier_detect=['Daily_Usage_Hours','Exercise_Hours','Screen_Time_Before_Bed','Weekend_Usage_Hours']
```

```python
outlier_detect=['Daily_Usage_Hours','Exercise_Hours','Screen_Time_Before_Bed','Weekend_Usage_Hours']

for i in outlier_detect:
  percentile_25=df[i].quantile(0.25)
  percentile_75=df[i].quantile(0.75)
  iqr=percentile_75-percentile_25
  upper_limit=percentile_75+1.5*iqr
  lower_limit=percentile_25-1.5*iqr
  percent_values_75=((df[i]>upper_limit).sum()/df.shape[0])*100
  percent_values_25=((df[i]<lower_limit).sum()/df.shape[0])*100
  print(i,percent_values_25,percent_values_75)
```

```python
for i in outlier_detect:
  percentile_25=df[i].quantile(0.25)
  percentile_75=df[i].quantile(0.75)
  iqr=percentile_75-percentile_25
  upper_limit=percentile_75+(1.5*iqr)
  lower_limit=percentile_25-(1.5*iqr)

  df[i]=np.where(df[i]>upper_limit,upper_limit,
                   np.where(df[i]<lower_limit,lower_limit,df[i]))
```

```python
ohe=OneHotEncoder(drop='first',handle_unknown='ignore',sparse_output=False)
sc=StandardScaler()

```

```python
ohe_to_encode=['Gender','School_Grade','Phone_Usage_Purpose']
```

```python
train_ohe=ohe.fit_transform(X_train[ohe_to_encode])
test_ohe=ohe.transform(X_test[ohe_to_encode])
```

```python
train_ohe_concat=pd.DataFrame(train_ohe,columns=ohe.get_feature_names_out(['Gender','School_Grade','Phone_Usage_Purpose']))
test_ohe_concat=pd.DataFrame(test_ohe,columns=ohe.get_feature_names_out(['Gender','School_Grade','Phone_Usage_Purpose']))
```

```python
X_train.drop(columns=ohe_to_encode,inplace=True)
X_test.drop(columns=ohe_to_encode,inplace=True)
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train.shape
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten,BatchNormalization
```

```python

from keras.layers import *
from keras import regularizers
from keras import initializers
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model=Sequential()
model.add(Dense(18, input_dim=18,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(16,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(16,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1,activation='linear'))
```

```python

model.summary()
```

```python
model.compile(optimizer='Adam', loss='mse', metrics=['r2_score'])
```

```python
callback = EarlyStopping(
    monitor="val_loss",
    min_delta=0.00001,
    patience=3,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False
)
```

```python
history=model.fit(X_train,y_train,epochs=100,validation_data=(X_test,y_test),callbacks=callback)
```

```python
y_hat=model.predict(X_test)
```

```python
r2_score(y_test,y_hat)
```
